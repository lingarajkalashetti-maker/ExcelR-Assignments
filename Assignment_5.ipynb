```python
## Outlier detection
## Boxplot
for col in df.columns:          # I am using this step to find out about/ Identify outliers or extreme values in my given data-set
    sns.boxplot(data=df,y=col) 
    plt.show()
```

```python
#### outlier capping
def outlier_capping(df,column):
    Q1= df[column].quantile(0.25)   # I am using this step to find values/ extreme outliers so it could prevent misleading results and analysis 
    Q3= df[column].quantile(0.75)   #what does this code do?
    IQR= Q3-Q1                      #It finds and caps outliers for every numerical column in my dataset using IQR(interquartile Range) method.
    lower_extreme= Q1-1.5*IQR       #Q1->25th percentile, Q2->75th percentile, IQR->Calculates the distance between Q1&Q2 (50% middle of the given data) 
    upper_extreme= Q3+1.5*IQR       #lower_extreme-> Defines absolute lowest acceptable boundary line and as for Upper_extreme value its vice versa
    df[column]= df[column].apply(lambda x: lower_extreme if x<lower_extreme else upper_extreme if x>upper_extreme else x)
for col in df.select_dtypes(include=['int','float']).columns:
    outlier_capping(df,col)
```

```python
#1.Defining the corrected function
def outlier_capping(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_extreme= Q1-1.5*IQR       #Q1->25th percentile, Q2->75th percentile, IQR->Calculates the distance between Q1&Q2 (50% middle of the given data) 
    upper_extreme= Q3+1.5*IQR       #lower_extreme-> Defines absolute lowest acceptable boundary line and as for Upper_extreme value its vice versa
    
   # Clean, standard way to cap values between a min and max 
    df[column]= df[column].clip(lower_extreme, upper_extreme) 
    return df  #It will return the dataframe unlike the logic i used in above code

   #2.Applying it to my numerical column(using number automatically grabs all ints and floats safely)
numeric_columns = df.select_dtypes(include = ["number"]).columns
for col in numeric_columns:
       df = outlier_capping(
           df, col
       )  #Re-assigning df to save changes.
 #NOTE = the above code i am using from the online platform after comparing the logics it suggested me that it fixes the bugs, ensures my data actually updates, and it saves changes safely.
```
