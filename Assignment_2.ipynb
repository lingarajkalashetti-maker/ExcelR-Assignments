11.	Create a small DataFrame manually with 10 rows and columns: 'Name', 'Age', 'City', and 'Salary'. Then:
- Use .info(), .describe()
- Select 'Name' and 'City' columns
- Drop 'City' column
- Fill any missing values in 'Salary' column with the mean
- Remove any duplicate rows


```python
import pandas as pd 
import numpy as np
import warnings
warnings.filterwarnings('ignore')

data ={
       'name':['raj','veer','rajesh','smithi','navven','veena','jai','hrithik','virat','priya'],
       'age':[24,25,27,28,32,26,30,31,26,33],
       'city':['bangalore','mumbai','delhi','kolkata','pune','ranchi','kochi','hydrabad','bidar','bengal'],
       'salary':[25000,22000,27000,32000,np.nan,29000,38000,40000,26000,50000]
}
df = pd.DataFrame(data)
print(df.info())
print(df.describe())

name_city = df[['name','city']]
drop = df.drop(columns=['city'])
df['salary'] = df['salary'].fillna(df['salary'].mean())
df = df.drop_duplicates()
print("\nFinal clean DataFrame")
print(df)
```
