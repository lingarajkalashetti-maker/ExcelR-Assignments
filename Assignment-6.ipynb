## ESTIMATION AND CONFIDENTIAL INTERVALS

```python
import numpy as np
import scipy.stats as stats
```

```python
# A.Durability data in millions of characters
data = np.array([1.13, 1.55, 1.43, 0.92, 1.25, 1.36, 1.32, 0.85, 1.07, 1.48, 1.20, 1.33, 1.18, 1.22, 1.29])
#The numbers n = 15
n = len(data)
sample_mean = np.mean(data)
sample_std = np.std(data, ddof=1) # ddof=1 for sample standard deviation

print(f"Sample Mean: {sample_mean:.4f}")
print(f"Sample Standard Deviation: {sample_std:.4f}")
print(f"Sample Size: {n}")
```

##### A. Build 99% Confidence Interval Using Sample Standard DeviationAssuming the sample is representative of the population, construct a 99% confidence interval for the mean number of characters printed before the print-head fails using the sample standard deviation. Explain the steps you take and the rationale behind using the t-distribution for this task.

```python
confidence_level = 0.99
alpha = 1 - confidence_level
df = n - 1

# Critical t-value
t_critical = stats.t.ppf(1 - alpha/2, df)

# Margin of error
margin_of_error_t = t_critical * (sample_std / np.sqrt(n))

# Confidence Interval
ci_t = (sample_mean - margin_of_error_t, sample_mean + margin_of_error_t)

print(f"t-Critical Value: {t_critical:.4f}")
print(f"99% CI (t-distribution): {ci_t}")
```

#### Part (B): 99% Confidence Interval Using Known Population Standard Deviation


```python
# Given population standard deviation (sigma)
population_std = 0.2

# Critical z-value for 99% confidence level
# using stats.norm.ppf since population std dev is known
alpha = 1 - confidence_level
z_critical = stats.norm.ppf(1 - alpha / 2)

# Margin of error using population standard deviation
margin_of_error_z = z_critical * (population_std / np.sqrt(n))

# Confidence Interval calculation
ci_z = (sample_mean - margin_of_error_z, sample_mean + margin_of_error_z)

print(f"z-Critical Value: {z_critical:.4f}")
print(f"99% CI (z-distribution): {ci_z}")
```

## HYPOTHESIS TESTING

```python
import numpy as np
import scipy.stats as stats
```

```python
# 1. Data Provided
n = 25                               # Sample size
sample_mean = 3050                   # Observed mean weekly cost (Rs.)
x_units = 600                        # Number of units produced
alpha = 0.05                         # Significance level
```

```python
# Model parameters: W = 1000 + 5X
intercept = 1000
slope = 5
theoretical_mean = intercept + (slope * x_units)
```

```python
# Standard deviation of cost: sigma = slope * SD of units
# Based on the assignment: 5 * 25
sigma = 5 * 25 

print(f"Theoretical Mean (μ): Rs. {theoretical_mean}")
print(f"Sample Mean (x̄): Rs. {sample_mean}")
print(f"Standard Deviation (σ): {sigma}")
```

```python
# 2. State the Hypotheses
# Null Hypothesis (H0): μ <= 4000 (Costs are not higher than the model suggests)
# Alternative Hypothesis (H1): μ > 4000 (Costs are significantly higher)

# 3. Calculate the Test Statistic (z)
# Formula: z = (x̄ - μ) / (σ / sqrt(n))
standard_error = sigma / np.sqrt(n)
z_score = (sample_mean - theoretical_mean) / standard_error
```

```python
# 4. Determine Critical Value and P-value
z_critical = stats.norm.ppf(1 - alpha)
p_value = 1 - stats.norm.cdf(z_score)

print("-" * 30)
print(f"Test Statistic (z): {z_score:.4f}")
print(f"Critical Value (z_crit): {z_critical:.4f}")
print(f"P-value: {p_value:.4f}")
print("-" * 30)
```

```python
# 5. Make a Decision and Conclusion
if z_score > z_critical:
    print("Decision: Reject the Null Hypothesis (H0).")
    print("Conclusion: There is strong evidence to support the claim that weekly operating costs are higher.")
else:
    print("Decision: Fail to Reject the Null Hypothesis (H0).")
    print("Conclusion: There is no strong evidence to support the claim that weekly operating costs are higher.")
```
