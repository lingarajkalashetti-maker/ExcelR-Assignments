```python
numerical_columns = df.select_dtypes(include=['number'])
print(numerical_columns.columns)
```

```python
# Calculation of mean, median, mode and standard deviation

mean = numerical_columns.mean()
print("\nMean of the coulmns ")
print(mean)
median = numerical_columns.median()
print("\nMedian of the coulmns ")
print(median)
mode = numerical_columns.mode().iloc[0]
print("\nMode of the coulmns ")
print(mode)
standard_deviation = numerical_columns.std()
print("\nStandard-devaiation of the coulmns ")
print(standard_deviation)
```

```python
stats = numerical_columns.describe()
stats
```

```python
print("Descriptive Statistics for Numerical columns ")
summary = pd.DataFrame(stats)
summary['mode'] = mode
print(summary)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
# Identifying columns types

numerical_columns = df.select_dtypes(include=['number'])
categorical_columns = df.select_dtypes(include=['object','category']).columns
print(numerical_columns)
print(categorical_columns)
```

# Histogram

```python
# ploting histogram for each numerical columns

print("plotting histogram.. ")
for col in numerical_columns:
    plt.figure(figsize=(8,4))
    sns.histplot(df[col], kde=True, color='red')
    plt.title(f'distribution of {col}')
    plt.xlabel(col)
    plt.ylabel('Frequency')
    plt.show()
```

# Box plot

```python
# Identifying outliers and interqaurtile range

for col in numerical_columns:
    plt.figure(figsize=(8,4))
    sns.boxplot(x=df[col], color='green')
    plt.title(f'Boxplot of {col}')
    plt.show()
```

# Bar chart

```python
# visualizing frequency of categorical column

for col in categorical_columns:
    plt.figure(figsize=(8,4))
    sns.countplot(data=df, x=col, palette='viridis', order=df[col].value_counts().index)
    plt.title(f'Frequency of categories in {col}')
    plt.xticks(rotation=40)
    plt.ylabel('count')
    plt.show()
```
